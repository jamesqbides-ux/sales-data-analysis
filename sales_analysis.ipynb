# Sales Data Analysis
## Business Insights from Sales Data



## Objective

The goal of this project is to analyze sales data to identify trends, top-performing categories, and regional performance to support business decision-making.

## Data Loading and Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('train.csv')

# Ver datos
df.head()
```

## Data Cleaning

- Converted date columns to datetime format  
- Checked for missing values  
- Prepared data for analysis  

```python
# Convertir fechas
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Crear columnas
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year

# Verificar columnas
print(df[['Order Date', 'Month', 'Year']].head())
```

## Data Analysis

```python
sales_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(sales_category)
```

```python
sales_month = df.groupby('Month')['Sales'].sum()
print(sales_month)
```

```python
sales_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(sales_region)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

## Data Visualization

```python
# Categorías
sales_category.plot(kind='bar', title='Ventas por categoría')
plt.show()

# Mes
sales_month.plot(title='Ventas por mes')
plt.show()

# Región
sales_region.plot(kind='bar', title='Ventas por región')
plt.show()
```

## Key Insights

- Technology is the top-performing category in terms of revenue  
- Sales show a consistent positive trend over time  
- The West region generates the highest sales  

These findings suggest strong growth potential and opportunities for strategic focus.

## Conclusion

This analysis highlights how data can be used to identify trends and support business decisions. Focusing on high-performing categories and regions can help improve revenue and optimize strategies.
